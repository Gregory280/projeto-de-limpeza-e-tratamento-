### Etapa 1
### Tratamento de Valores Ausentes

```python
# Função que calcula o percentual de valores ausentes
def func_calc_percent_valores_ausentes(df):
    total_celulas = np.prod(df.shape)
    contagem_ausentes = df.isnull().sum()
    total_ausentes = contagem_ausentes.sum()
    print("O dataset possui", round(((total_ausentes/total_celulas) * 100), 2), "%", "de valores ausentes.")
```

```python
func_calc_percent_valores_ausentes(df)
```

```python
# Função para calcular valores ausentes por coluna
def func_calc_percent_valores_ausentes_coluna(df):
    # Total de valores ausentes
    contagem_ausentes = df.isnull().sum() # mis_val

    # Porcentagem de valores ausentes
    contagem_ausentes_percentagem = 100 * contagem_ausentes / len(df)

    # Tipo de dado das colunas com valores ausentes
    contagem_ausentes_dtype = df.dtypes

    # Cria uma tabela com os resultados
    tabela_contagem_ausentes = pd.concat([contagem_ausentes, contagem_ausentes_percentagem, contagem_ausentes_dtype], axis=1)

    # Renomear as colunas
    tabela_contagem_ausentes.rename(columns = {0 : 'Valores Ausentes', 1 : '% de Valores Ausentes', 2: 'Dtype'}, inplace=True)

    # Classifica a tabela por porcentagem de valores ausentes de forma decrescente e remove colunas sem valores faltantes
    tabela_contagem_ausentes = tabela_contagem_ausentes[tabela_contagem_ausentes.iloc[:,0] != 0].sort_values(
        '% de Valores Ausentes', ascending = False).round(2)

    print ("O dataset possui " + str(df.shape[1]) + " colunas.\n"
        "Encontrado: " + str(tabela_contagem_ausentes.shape[0]) + " colunas que têm valores ausentes.")

    if tabela_contagem_ausentes.shape[0] == 0:
        return

    # Retorna o dataframe com informações ausentes
    return tabela_contagem_ausentes
```

```python
df_ausentes = func_calc_percent_valores_ausentes_coluna(df)
```

**Tratando os valores ausentes de colunas Categóricas**

Como são colunas categóricas utilizar ffill ou bfill não fazem sentido, pois colocar outros valores irá modificar a informação. Irei preencher os valores ausentes como desconhecido (unknown).

```python
def preencher_cat_desc(df, col, value):
    contagem = df[col].isna().sum()
    df[col] = df[col].fillna(value)
    if type(value) == 'str':
        print(f"{contagem} valores ausentes na coluna {col} foram substituídos por '{value}'.")
    else:
        print(f"{contagem} valores ausentes na coluna {col} foram substituídos por {value}.")
```

### Etapa 3 
### Tratamento de Outliers

A pedido de maior importancia, caso haja outliers nas colunas abaixo, serão substituidos pelos seus valores limites inferior e superior:\
Avg RTT DL (ms)\
Avg RTT UL (ms)\
Avg Bearer TP DL (kbps)\
Avg Bearer TP UL (kbps)  

```python
def substituir_outliers(df, coluna):
    qtd_antiga = df.shape[0]
    
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 2.5 * IQR
    limite_superior = Q3 + 2.5 * IQR

    contagem_inferior = df[coluna] < limite_inferior
    contagem_superior  = df[coluna] > limite_superior
    contagem_total = contagem_inferior.sum() + contagem_superior.sum()

    df.loc[df[coluna] < limite_inferior, coluna] = limite_inferior
    df.loc[df[coluna] > limite_superior, coluna] = limite_superior

    print(f"{contagem_total} valores da coluna '{coluna}' foram substituídos.")
```

```python
substituir_outliers(df_limpo, 'Avg RTT DL (ms)')
```

```python
substituir_outliers(df_limpo, 'Avg RTT UL (ms)')
```

```python
substituir_outliers(df_limpo, 'Avg Bearer TP DL (kbps)')
```

```python
substituir_outliers(df_limpo, 'Avg Bearer TP UL (kbps)')
```
